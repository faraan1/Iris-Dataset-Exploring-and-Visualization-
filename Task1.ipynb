```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv('Iris.csv')
print('Iris Dataset: ')
print(df)

print('Shape of Iris Dataset is: ')
print(df.shape)

print('Columns are: ') 
print(df.columns)

print('Value of Head is: ')      
print(df.head())

print('Information is: ')
print(df.info())

print('Describe: ')
print(df.describe())

fig,axes=plt.subplots(2,2,figsize=(12,10))
a1=axes[0,0]
a2=axes[0,1]
a3=axes[1,0]
a4=axes[1,1]

sns.scatterplot(data=df ,x='SepalLengthCm', y='SepalWidthCm', color='r',ax=a1)
a1.set_xlabel('SepalLengthCm')
a1.set_ylabel('SepalWidthCm')
a1.set_title('Scatter Plot')

sns.histplot(data=df['SepalLengthCm'],color='red',ax=a2)
a2.set_xlabel('SepalLengthCm')
a2.set_ylabel('Frequency')
a2.set_title('Histogram Plot')

sns.boxplot(data=df.iloc[:,1:5],ax=a3)
a3.set_xticklabels(df.columns[1:5],rotation=45)
a3.set_title('Box Plot')

axes[1,1].axis("off")

plt.show()
```
